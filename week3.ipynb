```python
import requests
from bs4 import BeautifulSoup
import csv
import os

# URL of the website to scrape
url = 'http://books.toscrape.com/'

# Make a GET request to fetch the raw HTML content
response = requests.get(url)

# Ensure response encoding is correct
response.encoding = response.apparent_encoding  # Set encoding to apparent encoding

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all products
products = soup.find_all('article', class_='product_pod')

# Create a list to hold the data
data = []

# Loop through the products and extract the title and price
for product in products:
    title = product.h3.a['title']  # Title
    price = product.find('p', class_='price_color').text  # Price

    print(f'Title: {title}, Price: {price}')  # DEBUG: Print to verify data extraction

    # Append to data list
    data.append([title, price])

# DEBUG: Check if data is populated
print(f'Total products found: {len(data)}')  # Number of products found

# Save to CSV
csv_file_path = 'books_bs.csv'
with open(csv_file_path, 'w', newline='', encoding='utf-8') as f:  # Specify utf-8 encoding for the CSV
    writer = csv.writer(f)
    writer.writerow(['Title', 'Price'])
    writer.writerows(data)

# Check if the file is created successfully
if os.path.exists(csv_file_path):
    print(f"Data scraping completed. Check '{csv_file_path}' for results.")
else:
    print("Failed to create CSV file.")
```

```python
# Required Imports
import scrapy
from scrapy.crawler import CrawlerProcess
import pandas as pd

class BooksSpider(scrapy.Spider):
    name = 'books'
    start_urls = ['http://books.toscrape.com/']

    def parse(self, response):
        # Extract book titles and prices
        titles = response.css('article.product_pod h3 a::attr(title)').getall()
        prices = response.css('p.price_color::text').getall()

        # Combine into a DataFrame
        data = pd.DataFrame({'Title': titles, 'Price': prices})

        # Save DataFrame to CSV
        data.to_csv('books_scrapy.csv', index=False)

# Run the Scrapy spider
process = CrawlerProcess()
process.crawl(BooksSpider)
process.start()
```

```python
# Required Imports
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Set up the Chrome driver
!apt-get update  # to update ubuntu to correctly run apt install
!apt install -y chromium-chromedriver
import sys
sys.path.insert(0, '/usr/lib/chromium-browser/chromedriver')

options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run in headless mode
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Create a new instance of the Chrome driver
driver = webdriver.Chrome(options=options)  # Corrected: Remove 'chromedriver' argument

# Open the website
driver.get('http://books.toscrape.com/')
time.sleep(3)  # Allow time for the page to load

# Extract titles and prices
titles = driver.find_elements(By.CSS_SELECTOR, 'article.product_pod h3 a')
prices = driver.find_elements(By.CSS_SELECTOR, 'p.price_color')

# Store the data in lists
book_titles = [title.get_attribute('title') for title in titles]
book_prices = [price.text for price in prices]

# Create a DataFrame and save to CSV
data = pd.DataFrame({'Title': book_titles, 'Price': book_prices})
data.to_csv('books_selenium.csv', index=False)

# Close the browser
driver.quit()
```
